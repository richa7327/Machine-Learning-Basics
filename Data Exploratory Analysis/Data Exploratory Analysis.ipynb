3. Plot	min and max for actual_mean_temp, actual_min_temp and 	actual_max_temp using bar graph. Use blue for min and red for max.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data to plot
n_groups = 3
minimun=(data['actual_mean_temp'].min() , data['actual_min_temp'].min() , data['actual_max_temp'].min())
maximum=(data['actual_mean_temp'].max() , data['actual_min_temp'].max() , data['actual_max_temp'].max())

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, minimun, bar_width,alpha=opacity,color='b',label='Minimum Values')

rects2 = plt.bar(index + bar_width, maximum , bar_width, alpha=opacity, color='r', label='Maximunm Values')

plt.xlabel('Attributes')
plt.ylabel('Min and Max Values')
plt.title('Min and Max Values')
plt.xticks(index + bar_width, ('actual_mean_temp' , 'actual_min_temp' , 'actual_max_temp'))
plt.legend()
plt.tight_layout()
plt.show()

```

4. Plot	two separate histograms for actual_mean_temp for months of Aug 2014 and March 2015. Make key observations (put them as comments in jupyter notebook).

```python
data1 = data[data['date'].str.contains("2014-8")]
actual_mean_august = data1['actual_mean_temp']

# the histogram of the data
plt.hist(actual_mean_august, 10, density=1, facecolor='green', alpha=0.75)
plt.xlabel('Actual Mean Temperature')
plt.ylabel('Probability')
plt.title('Actual mean temperature for August 2014')
plt.grid(True)
plt.show()

#The range of actual mean temperature lies between 68 and 85.
#The most occuring values of actual mean temperature are 77 and 78.
```

```python
#Histogram for actual mean temperature of March 2015
data2 = data[data['date'].str.contains("2015-3")]
actual_mean_march = data2['actual_mean_temp']

# the histogram of the data
plt.hist(actual_mean_march, 10, density=1, facecolor='brown', alpha=0.75)
plt.xlabel('Actual Mean Temperature')
plt.ylabel('Probability')
plt.title('Actual mean temperature for March 2015')
plt.grid(True)
plt.show()

#The range of actual mean temperature lies between 35 and 70.
#The most occuring values of actual mean temperature are 56 and 62.
```

6. Find the correlation between the following pairs of attributes: actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. Make key observations.


```python
import seaborn as sns
```

```python
data['actual_mean_temp'].corr(data['actual_min_temp'])
```

```python
#Correlation between actual mean temp and actual min temp.

df1 = pd.DataFrame(data, columns = ['actual_mean_temp','actual_min_temp'])
corrmat1 = df1.corr()
print(corrmat1)
f, ax = plt.subplots(figsize =(8, 8))
sns.heatmap(corrmat1, ax = ax, cmap ="Greens", linewidths = 0.1)

#Each attribute is completely correlated to itself and highly correlated to each other.
```

```python
#Correlation between actual mean temp and actual max temp.

df2 = pd.DataFrame(data, columns = ['actual_mean_temp','actual_max_temp'])
corrmat2 = df2.corr()
print(corrmat2)
f, ax = plt.subplots(figsize =(9, 8))
sns.heatmap(corrmat2, ax = ax, cmap ="Blues", linewidths = 0.1)

#Each attribute is completely correlated to itself and highly correlated to each other.
```

7. Plot	two scatter plots between the following pairs of attributes: 		actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. Make key observations.

```python
#Scatter plot between actual_min_temp and actual_mean_temp

actual_mean = data['actual_mean_temp']
actual_min = data['actual_min_temp']
plt.scatter(actual_mean, actual_min, c='r', alpha=0.5)
plt.xlabel('Actual Mean Temperature')
plt.ylabel('Actual Min temperature')
plt.show()

#The corresponding value of actual min temp is approximately same the the particular actual mean temp
```

```python
#Scatter plot between actual_max_temp and actual_mean_temp

actual_mean = data['actual_mean_temp']
actual_max = data['actual_max_temp']
plt.scatter(actual_mean, actual_max, c='g', alpha=0.5)
plt.xlabel('Actual Mean Temperature')
plt.ylabel('Actual Max temperature')
plt.show()

#The values of both attributes are almost similar to each other and forms a straight line
```

8.  Make the correlation matrix plot and scatterplot matrix taking all temperature attributes in the dataset. 

```python
data_temp = pd.DataFrame(data, columns = ['actual_mean_temp','actual_min_temp','actual_max_temp','average_min_temp','average_max_temp','record_min_temp','record_max_temp'])

#Correlation Matrix
corr = data_temp.corr()
#print(corr)
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0,cmap='Blues',square=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

#The matrix is symmetrical, i.e. the bottom left of the matrix is the same as the top right.
#This is useful as we can see two different views on the same data in one plot.
#We can also see that each variable is perfectly positively correlated with each other in the diagonal line from top left to bottom right.
                                          
                                          
```

```python
# Scatterplot Matrix

from pandas.plotting import scatter_matrix
scatter_matrix(data_temp, alpha = 0.6, figsize =(11,11))
plt.show()

#Like the Correlation Matrix Plot, the scatterplot matrix is symmetrical.
#This is useful to look at the pair-wise relationships from different perspectives.
#Because there is little point oi drawing a scatterplot of each variable with itself, the diagonal shows histograms of each attribute.
```
