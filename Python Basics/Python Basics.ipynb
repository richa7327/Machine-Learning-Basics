Ques1: Write a python program (menu driven) that asks for an operation to be performed on two
input numbers. Operations are add, subtract, divide, multiply. The program should keep asking for
options on new numbers until user specifies explicitly to exit. Also handle the scenario of divide by
zero through exception handling in python.

```python
x = True
while x:
        num1= int(input("Enter first number: "))
        num2= int(input("Enter second number: "))
        op= input("Enter operation to be performed: ")
        if(op == '+'):
                print("Result is: " +str(num1+num2))
        elif (op == '-'):
                print("Result is: " +str(num1-num2))
        elif(op == '*'):
                print("Result is: " +str(num1*num2))
        elif(op == '/'):
                if(num2 == 0):
                    raise Exception('Division by zero not possible')
                else:
                    print("Result is: "+str(num1/num2))
        else:
            print('Invalid Operator')
        y= input("Perform again?")
        
        if(y == 'no'):
                x= False

```

Ques2: Write a python program that takes a sentence as input from user and reverses the words in
the sentence. Eg. If the sentence is 'Welcome to AI lab', then the output should be 'lab AI to
Welcome'. (Use list data structure)

```python
sent_out = " "
sent = input("Enter a sentence: ")
sent_list = sent.split()
list_out = sent_list[::-1]
sent_out = sent_out.join(list_out)
print("Reversed sentence is: " +sent_out)
```

Ques4: Write a python program to find square root of a given number (n) within a given precision (p).
Do not use any predefined library. (Take a guess, compute error at each successive step, until error is
less than given precision)

```python
def squareRoot(n, p):
        start = 0;
        end = n;
        guess = (start+end)/2
        err = abs(guess**2 - n)
        while(err > p):
            if ((guess**2) > n):
                end = guess
            elif ((guess**2) < n):
                start = guess
            else:
                return guess
            guess = (start + end)/2
            err = abs(guess**2 - n)
        return guess
n = int(input("Enter the number: "))
p= float(input("Enter precision: "))
result = squareRoot(n, p)
print("Square root is: "+ str(result))
```

Ques5: Find the minimum value of a function y = $(x+3)^2$ . During lab evaluation, demonstrate the
solution to this problem using hand-written plot. Write python program to find the minimum value
programmatically, use gradient descent algorithm.

```python
cur_x = 4
rate = 0.01
precision = 0.001
previous_step_size = 1 #
max_iters = 10000
iters = 0
df = lambda x: 2*(x+3)
while previous_step_size > precision and iters < max_iters:
        prev_x = cur_x
        cur_x = cur_x - rate * df(prev_x)
        previous_step_size = abs(cur_x - prev_x)
        iters = iters+1
        print("Iteration",iters,"\nX value is",cur_x)
        
print("The local minimum occurs at", cur_x)
```
